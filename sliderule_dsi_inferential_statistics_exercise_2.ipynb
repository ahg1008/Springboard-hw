```python
CVT is applicable as the sample size is large enough.
 Null Hypothesis : The proportion of white sounding resumes and black sounding resumes that got a call is same.
A Z-test for proportions will be used.  
```

```python
w = data[data.race=='w']
b = data[data.race=='b']


```

```python
# Your solution to Q3 here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import scipy.stats as stats
import pylab



# white-sounding names
w_r = np.sum(w.call)
w_n = len(w)
w_p = (w_r / w_n)


# black-sounding names
b_r = np.sum(b.call)
b_n = len(b)
b_p = (b_r / b_n)







```

```python
#frequentist Approach

def ztest_proportions_two_samples(a1, b1, a2, b2, one_sided=False):
    
    p1 = a1/b1
    p2 = a2/b2
    
    p = (a1+a2)/(b1+b2)
    se = np.sqrt(p*(1-p)*(1/b1+1/b2))
    
    z = (p1-p2)/se
    p = 1-stats.norm.cdf(abs(z))
    p *= 2-one_sided
    return z, p
```

```python
# 95% confidence interval
prop_diff = w_p - b_p
print('Difference in proportions= \t {}\n'.format(prop_diff))

z_crit = 1.96
p_hat1 = w_p*(1-w_p)/w_n
p_hat2 =  b_p*(1-b_p)/b_n
ci_high = prop_diff + z_crit*(np.sqrt(p_hat1 + p_hat2))
ci_low = prop_diff - z_crit*(np.sqrt(p_hat1 + p_hat2))

z_stat, p_val = ztest_proportions_two_samples(w_r, w_n, b_r, b_n)
print('z-stat= \t {}\np-value: \t {}'.format(z_stat, p_val))

print('95% conf int: \t {} - {}'.format(ci_low, ci_high))
moe = (ci_high - ci_low)/2
print('Margin of err= \t +/-{}'.format(moe))
```

```python
#Bootstrap 


n = len(data)
r=r = np.sum(data.call)

callbacks = np.array([True] * int(r) + [False] * int(n-r))

size = 10000

bs_reps_diff = np.empty(size)

for i in range(size):
    w_bs_replicates = np.sum(np.random.choice(callbacks, size=w_n))
    b_bs_replicates = np.sum(np.random.choice(callbacks, size=b_n))
    
    bs_reps_diff[i] = (w_bs_replicates - b_bs_replicates)/b_n
    
bs_p_value = np.sum(bs_reps_diff >= prop_diff) / len(bs_reps_diff)

bs_ci = np.percentile(bs_reps_diff, [2.5, 97.5])
bs_mean_diff = np.mean(bs_reps_diff)

print('observed difference=  {}\n'.format(prop_diff))
print('BOOTSTRAP RESULTS\np-value: {}\n95% conf. int.: {}'.format(bs_p_value, bs_ci))
```
